```python
from astropy import wcs
# Create relevant FITS files
#First make Dendrogram
data, header = getdata(path+'column_properunits_conv36_source_only.fits', header=True)
wcs = wcs.WCS(header)
d = astrodendro.Dendrogram.compute(data, wcs=wcs, min_value=2e22, min_delta=5e22, min_npix=10)

# Make masks
hdu = fits.open(path+'column_properunits_conv36_source_only.fits')[0]

#############
# SgrB2
# Find the structures
sgrb2_17 = d[17]

# Extract the masks
mask_sgrb2 = sgrb2_17.get_mask()
mask_sgrb2[mask_sgrb2==True]=1
mask_sgrb2[mask_sgrb2==False]=0

# Create FITS HDU objects to contain the masks
mask_hdu_sgrb2 = fits.PrimaryHDU(mask_sgrb2.astype('short'), hdu.header)
mask_hdu_sgrb2.writeto(path+'dendromask_sgrb2.fits', clobber=True)

# Create masked column density map and save as FITS
sgrb2_column=hdu.data*mask_sgrb2
sgrb2_hdu_column=fits.PrimaryHDU(sgrb2_column, hdu.header)
sgrb2_hdu_column.writeto(path+'dendro_sgrb2.fits', clobber=True)

#############
# R > 100 pc
# Find the structures
rgt100_29 = d[29]
rgt100_23 = d[23]

# Extract the masks
mask_rgt100_29 = rgt100_29.get_mask()
mask_rgt100_23 = rgt100_23.get_mask()

#combine to make one mask
mask_rgt100 = mask_rgt100_29+mask_rgt100_23
mask_rgt100[mask_rgt100==2]=1

# Create FITS HDU objects to contain the masks
mask_hdu_rgt100 = fits.PrimaryHDU(mask_rgt100.astype('short'), hdu.header)
mask_hdu_rgt100.writeto(path+'dendromask_rgt100pc.fits', clobber=True)

# Create masked column density map and save as FITS
rgt100_column=hdu.data*mask_rgt100
rgt100_hdu_column=fits.PrimaryHDU(rgt100_column, hdu.header)
rgt100_hdu_column.writeto(path+'dendro_rgt100pc.fits', clobber=True)

#############
# R < 100 pc
# Find the structures
rlt100_19 = d[19]
rlt100_50 = d[50]
rlt100_53 = d[53]

# Extract the masks
mask_rlt100_19 = rlt100_19.get_mask()
mask_rlt100_50 = rlt100_50.get_mask()
mask_rlt100_53 = rlt100_53.get_mask()

#combine to make one mask
mask_rlt100 = mask_rlt100_19+mask_rlt100_50+mask_rlt100_53
mask_rlt100[mask_rlt100==2]=1

# Create FITS HDU objects to contain the masks
mask_hdu_rlt100 = fits.PrimaryHDU(mask_rlt100.astype('short'), hdu.header)
mask_hdu_rlt100.writeto(path+'dendromask_rlt100pc.fits', clobber=True)

# Create masked column density map and save as FITS
rlt100_column=hdu.data*mask_rlt100
rlt100_hdu_column=fits.PrimaryHDU(rlt100_column, hdu.header)
rlt100_hdu_column.writeto(path+'dendro_rlt100pc.fits', clobber=True)
```
